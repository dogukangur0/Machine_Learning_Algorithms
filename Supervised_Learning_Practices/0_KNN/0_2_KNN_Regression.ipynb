```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
X = np.sort(5*np.random.rand(40, 1), axis = 0)
y = np.sin(X).ravel() # flatten

y[::5] += 1 * (0.5 - np.random.rand(8)) # add noise

T = np.linspace(0, 5 , 500)[:, np.newaxis]
# first -> T.shape (500,) -> after T.shape (500, 1)

for i, weight in enumerate(["uniform", "distance"]):
  knn = KNeighborsRegressor(n_neighbors=5, weights = weight)
  y_pred = knn.fit(X, y).predict(T)

  plt.subplot(2, 1, i+1)
  plt.scatter(X, y, c = "green", label = "data")
  plt.plot(T, y_pred, c = "blue", label = "prediction")
  plt.title(f"KNN Regression weights : {weight}")
  plt.axis("tight")
  plt.legend()

plt.tight_layout()
plt.show()
```
