```python
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
```

```python
X, _ = make_blobs(n_samples = 300, n_features = 2, centers = 4, cluster_std = 0.6)
```

```python
plt.figure()
plt.scatter(X[:,0], X[:,1])
plt.title("Example Data")
```

```python
linkage_methods = ["ward", "single", "average", "complete"]

plt.figure(figsize = (15, 6))
for i, method in enumerate(linkage_methods):
  agg = AgglomerativeClustering(n_clusters=4, linkage=method)
  y_pred = agg.fit_predict(X)
  plt.subplot(2, 4, i+1)
  plt.title(f"{method.capitalize()} Linkage Dedrogram")
  dendrogram(linkage(X, method), no_labels=True)
  plt.xlabel("Data Points")
  plt.ylabel("Distance")

  plt.subplot(2, 4, i+5)
  plt.scatter(X[:, 0], X[:, 1], c = y_pred, cmap = "viridis")
  plt.title(f"{method.capitalize()} Linkage Clustering")
  plt.xlabel("X")
  plt.ylabel("y")
```
